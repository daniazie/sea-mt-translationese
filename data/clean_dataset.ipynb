```python
import json
import os
```

```python
dataset_dir = "../evaluation/predictions/baseline"
data_files = os.listdir(dataset_dir)
```

```python
files = {}
for data_file in data_files:
    if 'cleaned' in data_file:
        continue
    file_dir = dataset_dir + '/' + data_file
    with open(file_dir, "r") as file:
        files[file_dir] = json.load(file)
    # data_dir = dataset_dir + '/' + model_dir
    # splits = os.listdir(data_dir)
    # for split in splits:
    #     if 'cleaned' in split:
    #         continue
    #     split_dir = data_dir + '/' + split
    #     data_files[split_dir] = []
    #     with open(split_dir, "r") as file:
    #         for line in file.readlines():
    #             data_files[split_dir].append(json.loads(line))
    
```

```python
for file_name, data in files.items():
    processed = []
    for item in data:
        mt = item['mt']
        if not mt.startswith("###"):
            mt = mt.split('\n')[0].strip()
        else:
            if "### Malay:" in mt:
                mt = mt.split("### Malay:")[-1].strip()

        processed.append({
            "source": item['src'],
            "foreignization": mt,
            "domestication": item['ref']
        })

    with open(file_name.replace(".json", "_cleaned.jsonl"), "w") as file:
        for item in processed:
            json.dump(item, file)
            file.write('\n')
```
