```python
from pathlib import Path
import random
import json
import os
```

```python
results_dir = "../../evaluation/results"
```

```python
results = os.listdir(results_dir)
results
```

```python
random.seed(42)
for result in results:
    with open(f"{results_dir}/{result}", "r") as file:
        res = json.load(file)
    for r in res:
        r['mt'] = r['mt'].strip().splitlines()[0]
    with open(f"{results_dir}/{result}", "w") as file:
        json.dump(res, file, indent=2)
    res = random.sample(res, k=300)

    os.makedirs("/data/dania/sea-mt/data/gen_results", exist_ok=True)
    with open(f"/data/dania/sea-mt/data/gen_results/{result}l", "w") as file:
        for item in res:
            json.dump(item, file)
            file.write('\n')
```

```python
t_index_dir = "/data/dania/sea-mt/data/t-index_data/synthetic/enms"
for data_dir in os.listdir(t_index_dir):
    if Path(f"{t_index_dir}/{data_dir}/valid.jsonl").exists():    
        with open(f"{t_index_dir}/{data_dir}/valid.jsonl", 'r') as file:
            data = []
            for line in file.readlines():
                data.append(json.loads(line))
    elif Path(f"{t_index_dir}/{data_dir}/valid.json").exists():
        with open(f"{t_index_dir}/{data_dir}/valid.json", 'r') as file:
            data = json.load(file)

        proc_data = [
            {
                "src": item['source'],
                "mt": item['foreignization'],
                "ref": item['domestication']
            }
            for item in data
        ]

        with open(f"{t_index_dir}/{data_dir}/validation.jsonl", "w") as file:
            for item in proc_data:
                json.dump(item, file)
                file.write('\n')
```
