```python
import json
import os
```

```python
scores_dir = 'baseline'
scores = os.listdir(scores_dir)
scores.sort()
```

```python
MAP_MODELS = {
    "apertus-sealion-8b-it": "Apertus-SEA-LION-v4-8B-IT",
    "gemma-sealion-27b-it": "Gemma-SEA-LION-v4-27B-IT",
    "qwen-sealion-32b-it": "Qwen-SEA-LION-v4-32B-IT",
    "qwen35-27b": "Qwen3.5-27B",
    "sailor-20b": "Sailor2-20B",
    "translategemma-27b-it": "TranslateGemma-27B-IT"
}

MAP_SMALL_MODELS = {
    "apertus-sealion-8b-it": "Apertus-SEA-LION-v4-8B-IT",
    "gemma-sealion-4b-vl": "Gemma-SEA-LION-v4-4B-VL",
    "qwen-sealion-8b-vl": "Qwen-SEA-LION-v4-8B-VL",
    "qwen35-9b": "Qwen3.5-9B",
    "sailor-8b-chat": "Sailor2-8B-Chat",
    "translategemma-4b-it": "TranslateGemma-4B-IT"
}

MAP_DATA = {
    "ALT": "Parallel Asian Treebank",
    "ntrex-128": "NTREX-128"
}

MAP2MODELS = {
    v: k for k, v in MAP_MODELS.items()
}

MAP2DATASET = {
    v: k for k, v in MAP_DATA.items()
}
```

```python
metrics = ['Comet', 'XCOMET', 'CometKiwi', 'spBLEU', 'METEOR', '1 - T-index']
```

```python
def generate_table(ds):
    ind = 0
    for i, score_file in enumerate(scores):
        model, dataset, _, _ = score_file.split('_')
        if ds.lower() != dataset.lower():
            continue

        with open(f"{scores_dir}/{score_file}", "r") as file:
            results = json.load(file)


        if ind == 0:
            head = f"\\multirow{{2}}{{*}}{{Model}}\t&\t\\multicolumn{{3}}{{c}}{{Semantic-based}}\t&\t\\multicolumn{{2}}{{c}}{{N-gram-based}}\t&\t\\multirow{{2}}{{*}}{{1 - T-index}}\t&\t\\multirow{{2}}{{*}}{{Average}}\t\\\\"
            print(head)
            header = ""
            for metric in metrics:
                if not metric in head:
                    header += f"&\t{metric}\t"
                else:
                    header += "&\t"
            header += '&\t\\\\'
            print(header)
            print('\\midrule')

        row = f"{MAP_SMALL_MODELS[model]}\t"
        
        total = 0
        for metric in metrics:
            score = results[metric]
            if isinstance(score, dict):
                score = score.pop("meteor")
            if score < 1:
                score *= 100
            total += score
            row += f"&\t{score:.2f}\t"
        avg = total / len(metrics)
        row += f'&\t{avg:.2f}\t\\\\'

        print(row)
        ind += 1
```

```python
generate_table("alt")
```

```python
generate_table("ntrex-128")
```
